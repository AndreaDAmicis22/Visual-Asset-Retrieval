## 🧪 Query multiple — confronto rapido
Utile per valutare la qualità dell'encoder su più scenari.

```python
TEST_QUERIES = [
    # --- Oggetti Singoli e Animali (Baseline) ---
    "a person riding a bicycle",
    "a black cat sitting on a wooden chair",
    "a yellow school bus parked on the street",
    "an elephant standing next to a tree",
    # --- Relazioni Spaziali e Interazioni ---
    "a person holding an umbrella in the rain",
    "a pizza on a table next to a glass of wine",
    "a laptop computer sitting on top of a desk",
    "a bird perched on a traffic light",
    # --- Attributi e Colori (Specificità) ---
    "a blue suitcase at an airport",
    "a red stop sign with graffiti on it",
    "a white sheep in a green grassy field",
    "a man wearing a red tie and a suit",
    # --- Scene Complesse (Ragionamento) ---
    "a group of people playing frisbee in a park",
    "a kitchen with stainless steel appliances",
    "a busy city street with many cars and pedestrians",
    "a bathroom with a white sink and a large mirror",
    # --- Casi "Difficili" (Conteggio e Negazione) ---
    "two dogs running on the beach",
    "a street sign with no words on it",
    "a small teddy bear sitting on a large bed",
]

for q in TEST_QUERIES:
    visual_search(q, top_k=3, cols=3)
```
